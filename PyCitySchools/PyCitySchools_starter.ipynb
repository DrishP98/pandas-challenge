```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# total number of schools
total_schools = len(school_data_complete["school_name"].unique())
total_schools

# total number of students
total_students = school_data_complete["Student ID"].count()
total_students

# Calculate the Total Budget
total_budget = school_data["budget"].sum()
total_budget

# Calculate the Average Maths score
avg_math_score = round(school_data_complete["maths_score"].mean(),2)
avg_math_score

#calculate the average reading score
avg_reading_score = round(school_data_complete["reading_score"].mean(),2)
avg_reading_score

#Calculate the percentage of students with a passing maths score (50 or greater)
students_pass_math = student_data[student_data["maths_score"] >= 50]["student_name"].count()
percent_students_pass_math = round((students_pass_math/total_students)*100,0)
percent_students_pass_math

#Calculate the percentage of students with a passing reading score (50 or greater)
students_pass_reading = student_data[student_data["reading_score"] >= 50]["student_name"].count()
percent_students_pass_reading = round((students_pass_reading/total_students)*100,0)
percent_students_pass_reading

#Calculate the percentage of students who passed maths and reading (% Overall Passing)
overall_passing = (percent_students_pass_math + percent_students_pass_reading)/2
overall_passing

#Create a dataframe to hold the above results
area_summary = pd.DataFrame({
                    "Total Schools":[total_schools],
                      "Total Students":[total_students],
                      "Total Budget":[total_budget],
                      "Average Math Score":[avg_math_score],
                      'Average Reading Score':[avg_reading_score],
                      "% Passing Math":[percent_students_pass_math],
                      "% Passing Reading":[percent_students_pass_reading],
                      "% Overall Passing":[overall_passing]
                          })
area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_data_complete.head()
```

```python
# Determine the school type
school_type = school_data.set_index("school_name")["type"]
school_type

# total students
total_students_per_school = school_data_complete["school_name"].value_counts()
total_students_per_school

#total budget per school
budget_per_school = school_data_complete.groupby(["school_name"])["budget"].mean()
budget_per_school

#budget per student 
budget_per_student = budget_per_school/total_students_per_school
budget_per_student

#Average math Score
avg_math_scores = round(school_data_complete.groupby(["school_name"])["maths_score"].mean(),2)
avg_math_scores

#avergae reading scores
avg_reading_scores = round(school_data_complete.groupby(["school_name"])["reading_score"].mean(),2)
avg_reading_scores

# % passing maths
students_passing_math = school_data_complete[(school_data_complete["maths_score"]>=50)]
students_passing_math_per_school = students_passing_math.groupby(["school_name"])["student_name"].count()
percent_passing_math_schools = round((students_passing_math_per_school/total_students_per_school)*100,2)
percent_passing_math_schools

# % passing reading
students_passing_reading = school_data_complete[school_data_complete["reading_score"]>=50]
students_passing_reading_per_school = students_passing_reading.groupby(["school_name"])["student_name"].count()
percent_passing_reading_school = round((students_passing_reading_per_school/total_students_per_school)*100,2)
percent_passing_reading_school

# % Overall Passing (The percentage of students that passed maths and reading)
overall_passing_school = round((percent_passing_math_schools + percent_passing_reading_school)/2,2)
overall_passing_school

# Create a dataframe to hold the above results
per_school_summary = pd.DataFrame({"School Type":school_type,
                                 "Total Students":total_students_per_school,
                                 "Total School Budget":budget_per_school,
                                  "Per Student Budget":budget_per_student,
                                 "Average Maths Score":avg_math_scores,
                                 "Average Reading Score":avg_reading_scores,
                                 "% Passing Maths":percent_passing_math_schools,
                                 "% Passing Reading":percent_passing_reading_school,
                                 "% Overall Passing":overall_passing_school})
per_school_summary


```

```python
#Formatting to insert dollar sign
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

#Display
per_school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#create bins in which data will be held
bins = [0, 585, 630, 645, 675]
names = ["<$585", "$585-630", "$630-645", "$645-680"]

#categorize the spendings 
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = names)

#group by spending
by_spending = school_data_complete.groupby('spending_bins')


avg_math = by_spending['maths_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100
pass_read = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100
overall = (pass_math + pass_read) / 2

# create a dataframe         
spending_summary = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "% Overall Passing": overall})

spending_summary.index.name = "Spending Ranges (Per Student)"

#formating
spending_summary.style.format({'Average Math Score': '{:.2f}', 
                              'Average Reading Score': '{:.2f}', 
                              '% Passing Math': '{:.2f}', 
                              '% Passing Reading':'{:.2f}', 
                              '% Overall Passing': '{:.2f}'})



```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# create size bins
bins = [0, 1000, 2000,5000]
group_name = ["Small (<1000)", "Medium (1000-2000)" , "Large (2000-5000)"]

#categorize the spendings 
school_data_complete['size_bins'] = pd.cut(school_data_complete['size'], bins, labels = group_name)

#group by size
by_size = school_data_complete.groupby('size_bins')

#calculations 
average_math_score = by_size['maths_score'].mean()
average_reading_score = by_size['maths_score'].mean()
pass_maths_percent = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()*100
pass_reading_percent = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()*100
overall = (pass_maths_percent + pass_reading_percent) / 2
        
# create a dataframe           
size_summary = pd.DataFrame({
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    '% Passing Math': pass_maths_percent,
    '% Passing Reading': pass_reading_percent,
    '% Overall Passing': overall})
            

size_summary.index.name = "School Size"

#formating
size_summary.style.format({'Average Math Score': '{:.6f}', 
                              'Average Reading Score': '{:.6f}', 
                              '% Passing Math': '{:.6f}', 
                              '% Passing Reading':'{:.6f}', 
                              '% Overall Passing': '{:.6f}'})
```
